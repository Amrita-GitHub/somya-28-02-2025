```python
import pandas as pd
import matplotlib.pyplot as plt

sales_data = pd.read_csv('/content/drive/MyDrive/week9-files/sales-data.csv')
df = pd.DataFrame(sales_data)

sales_year_2023 = df[df['Year']==2023]
months_2023 = sales_year_2023['Month']
sales_2023 = sales_year_2023['Sales']
print(sales_year_2023)
# print(months_2023)
# print(sales_2023)

sales_year_2024 = df[df['Year']==2024]
months_2024 = sales_year_2024['Month']
sales_2024 = sales_year_2024['Sales']
print(sales_year_2024)
# print(months_2024)
# print(sales_2024)

fig, ax = plt.subplots(figsize=(6, 4))

# Add multiple lines with different styles
ax.plot(months_2024, sales_2024, color='blue', linestyle='-', marker='o',
        linewidth=2, markersize=8, label='2024')
ax.plot(months_2023, sales_2023, color='red', linestyle='--', marker='s',
        linewidth=2, markersize=8, label='2023')

# Add a legend
ax.legend(fontsize=12)

# Add title and labels
ax.set_title('Monthly Sales Comparison', fontsize=15)
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales (thousands $)', fontsize=12)

# Add grid for better readability
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# code using chatgpt
# ✅ Ensures months are sorted properly
# ✅ Formats the y-axis as currency
# ✅ Fixes the sales unit inconsistency
# ✅ Enhances plot readability with better size and labels

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
sales_data = pd.read_csv(r'C:\week9-files/sales-data.csv')

# Convert 'Month' to categorical for proper ordering
month_order = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
sales_data["Month"] = pd.Categorical(sales_data["Month"], categories=month_order, ordered=True)

# Filter data by year and sort
sales_2023 = sales_data[sales_data['Year'] == 2023].sort_values("Month")
sales_2024 = sales_data[sales_data['Year'] == 2024].sort_values("Month")

# Extract months and sales
months_2023, sales_values_2023 = sales_2023["Month"], sales_2023["Sales"]
months_2024, sales_values_2024 = sales_2024["Month"], sales_2024["Sales"]

# Create the plot
fig, ax = plt.subplots(figsize=(8, 5))

ax.plot(months_2024, sales_values_2024, color='blue', linestyle='-', marker='o',
        linewidth=2, markersize=8, label='2024 Sales')
ax.plot(months_2023, sales_values_2023, color='red', linestyle='--', marker='s',
        linewidth=2, markersize=8, label='2023 Sales')

# Formatting
ax.legend(fontsize=12)
ax.set_title('Monthly Sales Comparison (2023 vs 2024)', fontsize=15, fontweight='bold')
ax.set_xlabel('Month', fontsize=12)
ax.set_ylabel('Sales ($)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

# Format y-axis to show dollar amounts
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${x:,.0f}'))

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
