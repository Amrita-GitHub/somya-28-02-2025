```python
principle =1000
rate_of_intrest = 0.05
years = 5

amount = principle

for year in range(1, years+1):
  amount *= (1 + rate_of_intrest)
  print(f"for year : {year}   Compund Intrest: ${amount:.2f}")

```

```python
# print a range of values using the while loop

# 1. initialization

i = 0

# 2 check the condition, using a while loop

while i<10:

  # 3. process the data
  print(f"i= {i * 2}")

  # 4 Increment
  i += 1
```

```python
# print even/odd number using while loop
num = 0
while num < 10:
  if (num % 2 == 0 ):
    print(f"{num} is an Even Number")

  num += 1

```

```python
# usage of break - to terminate/exit the loop

for i in range(10):
  if (i == 5):
    break;
  print(f"i= {i}")
```

```python
# continue - ignore/bypass when the condition is true
for i in range(10):
  if (i == 5):
    continue;
  print(f"i= {i}")
```
