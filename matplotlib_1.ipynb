```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# prepare the data
# list
# x=[1,2,3,4,5]
# y=[2,4,6,8,10]

# numpy array
x = np.array([1,2,3,4,5])
y = np.array([2,4,6,8,10])

# create a line plot
# plt.plot(x,y,color='#FF0000',marker='o',linestyle=':')
# plt.plot(x,y,'o')

plt.subplot(1,2,1)
plt.plot(x,y,'s--g')


plt.subplot(1,2,2)
plt.plot(x,y,'o--r')


# add customizations
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.grid(True)
plt.legend(["Line 1"])

# display the plot
plt.show()
```
