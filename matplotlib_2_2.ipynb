```python
from google.colab import drive
# Mount Google Drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Load the data
product_performance = pd.read_csv('/content/drive/MyDrive/week9-files/product-performance.csv')
df = pd.DataFrame(product_performance)
```

```python
# Create a scatter plot showing the relationship between marketing spend and sales for each product
# Use point size to represent the profit margin
# Add a color gradient to represent ROI (Return on Investment)
# Include a trend line
# Add a title, axis labels, and a legend
```

```python
# Create a scatter plot showing the relationship between marketing spend and sales for each product
marketing_spend = df['Marketing_Spend']
units_sold= df['Units_Sold']
roi = df['ROI']
# scale profit margin
profit_margin = df['Profit_Margin']*10



fig, ax = plt.subplots(figsize=(7, 4))
# Create scatter plot with varying sizes based on campaign size
scatter = ax.scatter(marketing_spend, units_sold, s=profit_margin,
                    alpha=0.6, c=roi, cmap='viridis')

# Add a colorbar to indicate ROI
cbar = plt.colorbar(scatter)
cbar.set_label('ROI', fontsize=12)

# Add trend line
z = np.polyfit(marketing_spend, units_sold, 1)
p = np.poly1d(z)
ax.plot(marketing_spend, p(marketing_spend), "r--", alpha=0.8)



ax.set_title('Marketing Spend vs. Units Sold', fontsize=15)
ax.set_xlabel('Marketing Spend (thousands $)', fontsize=12)
ax.set_ylabel('Units Sold (thousands $)', fontsize=12)
ax.grid(True, linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
