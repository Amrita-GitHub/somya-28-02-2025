```python
from google.colab import drive
# Mount Google Drive
drive.mount('/content/drive')
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# Load the data
product_performance = pd.read_csv('/content/drive/MyDrive/week9-files/product-performance.csv')
df = pd.DataFrame(product_performance)
```

```python
# Create a pie chart showing the revenue distribution by product category
# Add percentage labels
# Explode the largest segment slightly
# Include a title and legend
```

```python
category = df['Category']
revenue = df['Revenue']

fig, ax = plt.subplots(figsize=(6, 4))

# Create pie chart
ax.pie(revenue, labels=category, autopct='%1.1f%%',
                                 startangle=140, shadow=True,
                                 colors=plt.cm.Paired.colors,
                                    explode=[0.1 if i == 0 else 0 for i in range(len(category))])


ax.set_title('revenue distribution by product category', fontsize=15)
ax.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

plt.tight_layout()
plt.show()
```
