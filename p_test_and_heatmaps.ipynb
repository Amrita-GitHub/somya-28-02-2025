```python
import numpy as np
from  scipy import stats

# !pip install scipy

# Sample sales data for two marketing campaigns
campaign_A = np.array([100, 102, 98, 105, 110, 95, 103])
campaign_B = np.array([95, 97, 96, 92, 90, 98, 94])

t_stats, p_val =  stats.ttest_ind(campaign_A, campaign_B)
print(t_stats)
print(p_val)
```

```python
# paired t-test

# Sales data before and after a new pricing strategy (paired observations)
sales_before = np.array([100, 102, 98, 105, 110, 95, 103])
sales_after  = np.array([105, 107, 100, 108, 115, 98, 105])
t_stats, p_val = stats.ttest_rel(sales_before, sales_after)
print(t_stats)
print(p_val)
```

```python
# corelation matrix
import pandas as pd


# Create sample data for correlation analysis
np.random.seed(42)
data = {
  'sales': np.random.normal(200, 50, 100),
  'marketing_spend': np.random.normal(50, 10, 100),
  'customer_satisfaction': np.random.uniform(1, 5, 100)
}

df = pd.DataFrame(data)
df

df_corr = df.corr()
df_corr


import seaborn as sns
sns.heatmap(df_corr, annot=True)

```
