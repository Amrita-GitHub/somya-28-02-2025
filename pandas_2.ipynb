```python
import pandas as pd
```

```python
# dictonary
data = {
   'product':['laptop', 'mouse','monitor','keyboard'],
   'price':[1000,50,75,100],
   'stock':[10,5,20,15]
}

# create a DataFrame from a dictonary
df = pd.DataFrame(data)
df
```

```python
data1 = [
    {
        'product':'laptop',
        'price':1000,
        'stock':10
    },
    {
        'product':'mouse',
        'price':50,
        'stock':5
    },
    {
        'product':'monitor',
        'price':75,
        'stock':20
    },
    {
        'product':'keyboard',
        'price':100,
        'stock':15
    }
]

df1 = pd.DataFrame(data1)
# df1 = pd.DataFrame(data1,index=['a','b','c','d'])
df1
```

```python
# retrieve column info
print(df1['product'])
```

```python
# retrieve rows
# loc(label based indexing) or iloc
print(df1.loc[:2])
# print(df1.loc['b'])
# slicing
# iloc-position based indexing
print(df1.iloc[:2])
```

```python
# filter on data
df1[df1['price']>=100]
```

```python
# Filter on strings
df1[df1['product'].str.contains('a')]
```

```python
# Preprocessing, cleaning and Transformation

data2 = [
    {
        'product':'laptop',
        'price':1000,
        'stock':10
    },
    {
        'product':'mouse',
        'price':None,
        'stock':5
    },
    {
        'product':'monitor',
        'price':75,
        'stock':20
    },
    {
        'product':'keyboard',
        'price':None,
        'stock':None
    },
    {
        'product':'pendrive',
        'price':None,
        'stock':None
    }
]

df2 = pd.DataFrame(data2)
# df2 = pd.DataFrame(data1,index=['a','b','c','d'])
df2
```

```python
# fill the entire dataset with all 0/default value
# df2.fillna(0,inplace=True)
# df2

# fill a specific column
df2['price'].fillna(0,inplace=True)
df2

# fill the avg for stock
df2['stock'].fillna(df2['stock'].mean(),inplace=True)
df2

# df2.fillna({'price': 0}, inplace=True)
# df2
```

```python
df2.dropna(inplace=True)
df2
```
