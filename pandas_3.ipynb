```python
import pandas as pd
```

```python
# Preprocessing, cleaning and Transformation

data2 = [
    {
        'product':'laptop',
        'price':1000,
        'stock':10
    },
    {
        'product':'mouse',
        'price':20,
        'stock':5
    },
    {
        'product':'monitor',
        'price':75,
        'stock':20
    },
    {
        'product':'keyboard',
        'price':50,
        'stock':10
    },
    {
        'product':'pendrive',
        'price':100,
        'stock':1
    },
    {
        'product':'pendrive',
        'price':None,
        'stock':None
    }
]

df2 = pd.DataFrame(data2)
# df2 = pd.DataFrame(data1,index=['a','b','c','d'])
df2
```

```python
df2.info()
```

```python
# check if the dataset contains duplicated rows
df2[df2.duplicated()]
```

```python
# drop the duplicated rows
df2.drop_duplicates
```

```python
# fill the entire datset with 0
df2.fillna(0,inplace=True)
df2
```

```python
# create a new column , total price
df2['Total Price'] = df2['price'] * df2['stock']
df2
```

```python
# create a column to show the discounted price
df2['Discounted Price'] = df2['price'] * 0.9
df2
```

```python
# rename columns
df2.rename(columns={'product':'Item', 'stock' : 'Inventory'},inplace=True)
df2
```

```python
# convert the float datatype to int datatype
df2['price'] = df2['price'].astype(int)
print(df2.info())
df2
```

```python
# group sales by region
sales_data = {
  'Region': ['North', 'South', 'East', 'West', 'North', 'South'],
  'Sales': [250, 300, 450, 500, 350, 400]
}
df3 = pd.DataFrame(sales_data)
df3
```

```python
# no. of sales in a region
grouped_sales = df3.groupby('Region').count()
grouped_sales
```

```python
# sum of a sales in a region
grouped_sales = df3.groupby('Region')['Sales'].sum()
grouped_sales
```

```python
# aggregate data using multiple functions

# grouped_sales = df3.groupby('Region')['Sales'].agg({'sum','mean','count'})
grouped_sales = df3.groupby('Region').agg({'Sales':['sum','mean','count']})
grouped_sales
```
