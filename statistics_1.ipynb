```python
import pandas as pd

# Sample sales data
data = {'sales': [150, 200, 250, 200, 210, 205, 175, 225, 275, 223, 209, 195, 192, 240, 237]}
df = pd.DataFrame(data)
df
print(df.describe())
print('----------------------------')
print(f'mean: {df.mean()}')
print(f'median: {df.median()}')
print(f'mode: {df.mode()}')
print(f'standard deviation: {df.std()}')
print(f'variance: {df.var()}')

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig,ax = plt.subplots(figsize=(7,4))
# ax.hist(df['sales'],bins=6,edgecolor='black')
sns.histplot(df['sales'],edgecolor='black',kde=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,ax = plt.subplots(figsize=(7,4))
# ax.boxplot(df['sales'])
sns.boxplot(df['sales'])
plt.show()
```
